```python
data["Tax_Income"] = pd.cut(data["Taxable.Income"], bins = [10002,30000,99620], labels = ["Risky", "Good"])
```

```python
data.head()
```

```python
data.tail(10)
```

```python
data = pd.get_dummies(data, columns = ["Tax_Income"], drop_first=True)
```

```python
data.head(20)
```

```python
def norm_fun(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
data_norm = norm_fun(data.iloc[:,1:])
```

```python
data_norm.head(10)
```

```python
x = data_norm.drop(['Tax_Income_Good'], axis = 1)
y = data_norm['Tax_Income_Good']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(criterion = 'gini', max_depth = None)
```

```python
model.fit(x_train, y_train)
```

```python
print('Train_Accuracy : {}'.format(model.score(x_train, y_train)))
print('Test_Accuracy : {}'.format(model.score(x_test, y_test)))
```

```python
model_new = RandomForestClassifier(criterion = 'gini', max_depth = 3)
```

```python
model_new.fit(x_train, y_train)
```

```python
print('Train_Accuracy : {}'.format(model_new.score(x_train, y_train)))
print('Test_Accuracy : {}'.format(model_new.score(x_test, y_test)))
```
